### Top-Selling Artists: Identify artists with the highest sales and analyze their sales trends over time

```python
%%sql
top_artists <<
SELECT 
    a.Name AS ArtistName,
    ROUND(SUM(ii.UnitPrice * ii.Quantity), 2) AS TotalSales
FROM 
    artists a
    JOIN albums al ON a.ArtistId = al.ArtistId
    JOIN tracks t ON al.AlbumId = t.AlbumId
    JOIN invoice_items ii ON t.TrackId = ii.TrackId
GROUP BY 
    a.Name
ORDER BY 
    TotalSales DESC
LIMIT 10;
```

```python
%%sql
WITH Top_Artists AS (
    -- Create a Common Table Expression (CTE) called 'Top_Artists' to calculate the top 10 artists based on total sales
    SELECT ar.ArtistId, ar.Name AS Artist, 
           ROUND(SUM(ii.Quantity * ii.UnitPrice), 2) AS Total_Sales
    FROM artists ar
    JOIN albums al ON ar.ArtistId = al.ArtistId
    JOIN tracks tr ON al.AlbumId = tr.AlbumId
    JOIN invoice_items ii ON ii.TrackId = tr.TrackId
    JOIN invoices i ON i.InvoiceId = ii.InvoiceId
    GROUP BY ar.ArtistId
    ORDER BY Total_Sales DESC
    LIMIT 10
)
-- Get sales trends for these top 10 artists by month
SELECT ar.Name AS Artist, 
       ROUND(SUM(ii.Quantity * ii.UnitPrice), 2) AS Total_Sales, 
       strftime('%Y-%m', i.InvoiceDate) AS Month
FROM artists ar
JOIN albums al ON ar.ArtistId = al.ArtistId
JOIN tracks tr ON al.AlbumId = tr.AlbumId
JOIN invoice_items ii ON ii.TrackId = tr.TrackId
JOIN invoices i ON i.InvoiceId = ii.InvoiceId
JOIN Top_Artists ta ON ar.ArtistId = ta.ArtistId
GROUP BY strftime('%Y-%m', i.InvoiceDate), ar.Name
ORDER BY ta.Total_Sales DESC, Month
;
```

# USE CASE 2 
### Customer Purchase Patterns: Segment customers based on purchase behavior and identify key characteristics of high-value customers.


```python
%%sql
-- Common Table Expression (CTE) to get the most recent invoice date
WITH LatestDate AS (
    SELECT MAX(InvoiceDate) AS MaxDate
    FROM invoices
),
-- CTE to calculate customer statistics such as total spending, purchase frequency, and recency of last purchase
CustomerStats AS (
    SELECT 
        c.CustomerId,
        ROUND(SUM(ii.UnitPrice * ii.Quantity), 2) AS Total_Spending,
        COUNT(DISTINCT i.InvoiceId) AS Purchase_Frequency,
        MAX(i.InvoiceDate) AS Last_Purchase_Date,
        ROUND(JULIANDAY((SELECT MaxDate FROM LatestDate)) - JULIANDAY(MAX(i.InvoiceDate)), 0) AS Recency
    FROM 
        customers c
        JOIN invoices i ON c.CustomerId = i.CustomerId
        JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
    GROUP BY 
        c.CustomerId
),
-- CTE to segment customers based on their total spending, recency, and purchase frequency
CustomerSegments AS (
    SELECT 
        cs.CustomerId,
        c.FirstName || ' ' || c.LastName AS CustomerName,
        c.Country,
        cs.Purchase_Frequency,
        cs.Total_Spending,
        cs.Last_Purchase_Date,
        cs.Recency,
        CASE 
            WHEN cs.Total_Spending > 45 AND cs.Recency <= 180 THEN 'VIP' -- High spenders who bought recently
            WHEN cs.Total_Spending > 45 AND cs.Recency > 180 THEN 'Churned VIP'  -- High spenders who have not bought recently
            WHEN cs.Total_Spending BETWEEN 40 AND 45 THEN 'Loyal' -- Loyal customers with good spending
            WHEN cs.Total_Spending BETWEEN 37 AND 40 THEN 'Regular' -- Regular customers with moderate spending
            WHEN cs.Recency > 180 THEN 'Churned' -- Customers who have not bought in the past 180 days
            ELSE 'New' -- New customers 
        END AS CustomerSegment
    FROM 
        CustomerStats cs
        JOIN customers c ON cs.CustomerId = c.CustomerId
)
SELECT * 
FROM CustomerSegments
ORDER BY Total_Spending DESC
```

### Monthly Sales (Year-Month)

```python
%%sql
SELECT 
    STRFTIME('%Y-%m', i.InvoiceDate) AS YearMonth,
    ROUND(SUM(ii.UnitPrice * ii.Quantity), 2) AS TotalSales
FROM 
    invoices i
    JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
GROUP BY 
    STRFTIME('%Y-%m', i.InvoiceDate)
ORDER BY 
    YearMonth
```

```python
# assign the result to
monthly_sales = _
```

```python
# Convert the result into a Pandas DataFrame
monthly_sales_df = monthly_sales.DataFrame()
```

```python
# Display the first few rows of the monthly sales DataFrame to get an overview of the data
monthly_sales_df.head()
```

```python
# Extract the 'Month' part from the 'YearMonth' column (last two characters) to focus on the month of each sales record
monthly_sales_df['Month'] = monthly_sales_df['YearMonth'].str[-2:]
```

```python
# Grouping the data by 'Month' to calculate the average sales per month
seasonal_df = monthly_sales_df.groupby('Month')['TotalSales'].mean().reset_index()
```

```python
# Calculate the mean and standard deviation of total sales to identify significant sales events
# The mean and standard deviation help to set a threshold for identifying sales events that are significantly higher than average
mean_sales = monthly_sales_df['TotalSales'].mean()
std_sales = monthly_sales_df['TotalSales'].std()

# Define a threshold that is 2 standard deviations above the mean to identify significant sales events
threshold = mean_sales + 2 * std_sales

# Filter the monthly_sales_df to find sales events where TotalSales are above the threshold, indicating unusually high sales
significant_events_df = monthly_sales_df[monthly_sales_df['TotalSales'] > threshold]
```

```python
# Display the outcome
print("Significant Sales Events (Above 2 Std Dev):")
print(significant_events_df)
```

# USE CASE 5


### Customer Lifetime Value (CLV): Calculate the lifetime value of customers based on their purchase history and provide recommendations for improving customer retention.

```python
%%sql
SELECT 
    c.CustomerId,
    c.FirstName || ' ' || c.LastName AS CustomerName,
    COUNT(DISTINCT i.InvoiceId) AS PurchaseCount,
    ROUND(SUM(ii.UnitPrice * ii.Quantity), 2) AS TotalSpent,
    MIN(i.InvoiceDate) AS FirstPurchase,
    MAX(i.InvoiceDate) AS LastPurchase,
    ROUND(JULIANDAY(MAX(i.InvoiceDate)) - JULIANDAY(MIN(i.InvoiceDate)), 0) AS TenureDays
FROM 
    customers c
    JOIN invoices i ON c.CustomerId = i.CustomerId
    JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
GROUP BY 
    c.CustomerId, c.FirstName, c.LastName
ORDER BY 
    TotalSpent DESC
```

```python
# Assign result to
clv = _
```

```python
# Convert result to a DataFrame
clv_df = clv.DataFrame()
```

```python
# Display first five rows
clv_df.head()
```

```python
# Recency calculation

# Assigning the max date to the last day of the last year in the chinook database
max_date = '2013-12-31'  

# Converting the 'LastPurchase' column to datetime format for accurate date calculations.
clv_df['LastPurchaseDate'] = pd.to_datetime(clv_df['LastPurchase'])
clv_df['RecencyDays'] = (pd.to_datetime(max_date) - clv_df['LastPurchaseDate']).dt.days

# defining a function, 'assign_segment', that classifies customers into segments based on their total spending (TotalSpent)
# and recency (RecencyDays)
def assign_segment(row):
    if row['TotalSpent'] > 45 and row['RecencyDays'] <= 180:
        return 'VIP'
    elif row['TotalSpent'] > 45 and row['RecencyDays'] > 180:
        return 'Churned VIP'
    elif 40 <= row['TotalSpent'] <= 45:
        return 'Loyal'
    elif 37 <= row['TotalSpent'] <= 40:
        return 'Regular'
    elif row['RecencyDays'] > 180:
        return 'Churned'
    else:
        return 'New'
```

```python
# Applying the 'assign_segment' function to each row in the dataframe to categorize customers into CLV segments.
clv_df['CLV_Segment'] = clv_df.apply(assign_segment, axis=1)
```

```python
# Printing the count of customers in each CLV segment to see the distribution of segments.
print(clv_df.groupby('CLV_Segment').size())
```

```python
# Displaying the first 10 rows of the dataframe with key customer information
print(clv_df[['CustomerName', 'TotalSpent', 'RecencyDays', 'CLV_Segment']].head(10))
```

```python
# Plotting a histogram to visualize the distribution of total spending (CLV) among customers

plt.figure(figsize=(10, 6))
plt.hist(clv_df['TotalSpent'], bins=20, color='purple', edgecolor='black')
plt.xlabel('Total Spent ($)')
plt.ylabel('Number of Customers')
plt.title('CLV Distribution')
plt.tight_layout()
plt.show()
```

```python
# Plotting a bar chart to visualize the top 10 customers based on their total spending (CLV)

plt.figure(figsize=(12, 6))
top_10 = clv_df.head(10)
plt.bar(top_10['CustomerName'], top_10['TotalSpent'], 
        color=top_10['CLV_Segment'].map({'VIP': 'gold', 'Churned VIP': 'orange', 'Loyal': 'blue', 'Regular': 'grey', 'Churned': 'red', 'New': 'green'}))
plt.xticks(rotation=45, ha='right')
plt.xlabel('Customer Name')
plt.ylabel('Total Spent ($)')
plt.title('Top 10 Customers by CLV')
plt.tight_layout()
plt.show()
```

```python
# Plotting a scatter plot to analyze the relationship between customer tenure (days) and total spending (CLV)
# Each point represents a customer, with the 'hue' differentiating customers by their CLV segment (e.g., VIP, Loyal, Churned)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=clv_df, x='TenureDays', y='TotalSpent', hue='CLV_Segment', size='PurchaseCount')
plt.xlabel('Tenure (Days)')
plt.ylabel('Total Spent ($)')
plt.title('CLV vs. Tenure by Segment')
plt.tight_layout()
plt.show()
```

```python
# Retention Recommendations 

# VIP (>45, Recent): Loyalty rewards.

# Churned VIP (>45, Inactive): Re-engagement campaigns.

# Loyal (40-45): Incentives to reach VIP status.

# Regular (37-40): Promotions to boost spending.

# Churned (<37, Inactive): Low-cost re-engagement.

# New (<37, Recent): Onboarding offers.
```

# USE CASE 6

### Market Basket Analysis: Perform association rule mining to find common combinations of tracks or albums purchased together.


```python
%%sql
basket << 
SELECT 
    i.InvoiceId,
    a.Title AS AlbumTitle
FROM 
    invoices i
    JOIN invoice_items ii ON i.InvoiceId = ii.InvoiceId
    JOIN tracks t ON ii.TrackId = t.TrackId
    JOIN albums a ON t.AlbumId = a.AlbumId
GROUP BY 
    i.InvoiceId, a.AlbumId
ORDER BY 
    i.InvoiceId, a.Title;
```

```python
# Convert to Dataframe
basket_df = basket.DataFrame()
```

```python
# Display first five rows
basket_df.head()
```

```python
# Creating a crosstab table (pivot table) where rows represent each InvoiceId and columns represent AlbumTitle
# The values are set to boolean (True/False), indicating whether a particular album was purchased in a given invoice

basket_albums = pd.crosstab(basket_df['InvoiceId'], basket_df['AlbumTitle']).astype(bool)
```

```python
# Applying the Apriori algorithm to the basket data to find frequent itemsets
# The min_support value of 0.008 was selected as the best balance between finding meaningful associations 
# and avoiding an excessive number of itemsets after testing various values for min_support 

frequent_itemsets = apriori(basket_albums, min_support=0.008, use_colnames=True)
```

```python
# Confidence directly measures the likelihood that a rule holds true, making it a straightforward way to assess the strength of associations

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules['item_count'] = rules['antecedents'].apply(len) + rules['consequents'].apply(len)

# Keep only pairs # 
#added (.copy) to avoid SettingwithCopyWarning
pair_rules = rules[rules['item_count'] == 2].copy()
```

```python
# Sorting the rules by 'lift' in descending order to identify the top 10 strongest associations
# Added (.copy) to avoid SettingwithCopyWarning

top_rules = pair_rules.sort_values(by='lift', ascending=False).head(10).copy() 
```

```python
# Formatting the 'itemsets', 'antecedents', and 'consequents' columns as comma-separated strings for easier readability and interpretation.

frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(lambda x: ', '.join(x))
pair_rules['antecedents'] = pair_rules['antecedents'].apply(lambda x: ', '.join(x))
pair_rules['consequents'] = pair_rules['consequents'].apply(lambda x: ', '.join(x))
top_rules['antecedents'] = top_rules['antecedents'].apply(lambda x: ', '.join(x))
top_rules['consequents'] = top_rules['consequents'].apply(lambda x: ', '.join(x))
```

```python
# Displaying the top 10 association rules based on the highest lift value.
print("Top 10 Album Association Rules by Lift:")
print(top_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].round(3).to_string(index=False))
```

```python
# Import warning to avoid deprecated function warning
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
# Plotting a scatterplot of the top 10 album association rules, visualizing support vs. confidence
# with the size of the markers representing the lift of each rule

plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_rules, x='support', y='confidence', size='lift', hue='lift', palette='viridis', sizes=(50, 500))
plt.title("Top 10 Album Association Rules by Lift")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.legend(title="Lift", loc='best')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

#### Spreading out since overlapping points stack

```python
# Add manual jitter to spread out the points
np.random.seed(42)  # For reproducibility
jitter_support = np.random.uniform(-0.0005, 0.0005, size=len(top_rules))  # Small wiggle for support
jitter_confidence = np.random.uniform(-0.02, 0.02, size=len(top_rules))  # Small wiggle for confidence
top_rules['support_jittered'] = top_rules['support'] + jitter_support
top_rules['confidence_jittered'] = top_rules['confidence'] + jitter_confidence

# Visualization with jittered values
plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_rules, x='support_jittered', y='confidence_jittered', size='lift', hue='lift', 
                palette='viridis', sizes=(50, 500), alpha=0.7)
plt.title("Top 10 Album Association Rules by Lift")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.legend(title="Lift", loc='best')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
